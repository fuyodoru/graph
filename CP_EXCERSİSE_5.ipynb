**The else clause**

```python
if x % 2 == 0:
    print('x is even')
else:
    print('x is odd')
```

```python
if x < y:
    print('x is less than y')
elif x > y:
    print('x is greater than y')
else:
    print('x and y are equal')
```

```python
if x == y:
    print('x and y are equal')
else:
    if x < y:
        print('x is less than y')
    else:
        print('x is greater than y')
```

```python
def countdown(n):
    if n <= 0:
        print('Blastoff!')
    else:
        print(n)
        countdown(n-1)
```

```python
countdown(5)
```

```python
def print_n_times(string, n):
    if n > 0:
        print(string)
        print_n_times(string, n-1)
```

```python
print_n_times('Batman ', 4)
```
